```python
import os
os.sys.path
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
img = cv2.imread("Crow.jpg")
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
r, g, b = cv2.split(img1)
r = r.flatten()
g = g.flatten()
b = b.flatten()
#plotting 
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(r, g, b)
plt.show()
```

```python
data_vector = img1.reshape((-1,3))
data_vector = np.float32(data_vector)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 2
attempts=10
ret,label,center=cv2.kmeans(data_vector,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img1.shape))
figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,2,1),plt.imshow(img1)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(result_image)
plt.title('Segmented Image when K = %i' % K), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def sampling(N):
    l=[]
    l=np.random.uniform(0,1,N)
    n1=len([i for i in l if (i<=0.35)])
    n2=len([i for i in l if (i >= 0.35)])
    
    return n1,n2
```

```python
n1,n2=sampling(1000)
print("No of Training samples in dataset - ", n1)
print("No of Training samples in dataset + ", n2)
```

```python
mean=[0,0]
covariance=[[1,0],[0,1]]
x1 = np.random.multivariate_normal(mean,covariance,n1)
x1.shape
```

```python
from matplotlib import pyplot as plt
import numpy as np

rho = np.sqrt(np.random.uniform(4, 9, n2))
phi = np.random.uniform(-np.pi, np.pi, n2)

x = rho * np.cos(phi)
y = rho * np.sin(phi)

plt.figure(figsize = (10,10))
#plt.scatter(x, y, s = 4)

x2=np.zeros((n2,2))
for i in range(n2):
    x2[i][0]=x[i]
    x2[i][1]=y[i]
    
```

```python
X=np.vstack((x1,x2))
```

```python
plt.figure(figsize = (10,10))
plt.scatter(x1[:,0],x1[:,1])
plt.scatter(x2[:,0],x2[:,1])
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Data Generation')
```

```python
y1=np.array([[0] for i in range(n1)])
y2=np.array([[1] for i in range(n2)])
Y=np.vstack((y1,y2))

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

tuned_parameters = [{'kernel': ['linear'], 'C': [0.001,0.01,0.1,1,10,100]}]
C=np.array([0.001,0.01,0.1,1,10,100])
clf = GridSearchCV(SVC(), tuned_parameters, cv=10)                       
clf.fit(X, Y.ravel())

print("Best parameters set found on Training Dataset:")
print()
print(clf.best_params_)
print()
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']

for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f  for %r" % (mean,  params))
print()

y_true, y_pred = Y, clf.predict(X)
#print((y_true, y_pred))
print("training accuracy",accuracy_score(y_true,y_pred))

```

```python
plt.plot(C,1-means)
plt.xlabel('C values')
plt.ylabel('error')
plt.title('C vs error')
```

```python
index1=[]
index2=[]
for i in range(n1):
    if y_pred[i]!=0:
        index1.append(i)
for i in range(n1,n2):
       if y_pred[i]!=1:
            index2.append(i)
error1=np.zeros((len(index1),2))
error2=np.zeros((len(index2),2))

for i in range(len(index1)):
    error1[i,:]=X[index1[i],:]
for i in range(len(index2)):
    error2[i,:]=X[index2[i],:]

print(error1.shape)
print(error2.shape)
plt.figure(figsize = (10,10))
plt.scatter(x1[:,0],x1[:,1],label='Dataset -')
plt.scatter(x2[:,0],x2[:,1],label='Dataset +')
plt.scatter(error1[:,0],error1[:,1],color='red',label='- classified as +')
plt.scatter(error2[:,0],error2[:,1],color='green',label='+ classified as -')
plt.legend(scatterpoints=3,loc='upper left',ncol=5,fontsize=10)
plt.ylabel('Y values')
plt.xlabel('X values')
plt.title(' Data after Linear SVM training ')
```
